```python
d= {} # {} , 딕션너리 의미 
d["x"] = 1
d["b"] = 2
d["d"] = 4
d["c"] = 3
d["e"] = 5

for k,v in d.items():
  print(f"key : {k}, value : {v}")
```

```python
# OrderdDict , 순서있는 딕션너리 이용하기. 
from collections import OrderedDict
d= OrderedDict()
d["x"] = 1
d["b"] = 2
d["d"] = 4
d["c"] = 3
d["e"] = 5

for k,v in d.items():
  print(f"key : {k}, value : {v}")
```
