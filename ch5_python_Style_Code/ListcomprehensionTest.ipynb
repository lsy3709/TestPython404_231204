```python
# 기존에 새리스트 추가시, 반복문이나, 안에 로직을 이용해서, 작업.

# 컴프리헨션으로 작업을 하면, 해당 리스트에 반복문을 주입해서, 바로 생성. 

result = [i for i in range(5)]
print(f"result 조회 : {result}")
```

```python
# 컴프리헨션으로 작업을 하면, 해당 리스트에 반복문을 주입해서, 바로 생성. 
# 조건문 추가하기. 
result = [i for i in range(5) if i % 2 ==0]
print(f"result 조회 : {result}")
```
